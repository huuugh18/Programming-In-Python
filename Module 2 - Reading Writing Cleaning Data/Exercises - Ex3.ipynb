```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Data/yeg_weather.csv', index_col=0, usecols=['Row ID', 'Minimum Temperature (C)', 'Date'])


# sorted_data = data[['Date', 'Minimum Temperature (C)']].sort_values(by='Minimum Temperature (C)')
sorted_data = data.sort_values(by='Minimum Temperature (C)')
# sorted_data

data.head()
```

```python
#get min temp
temp_min_col = sorted_data['Minimum Temperature (C)']

temp_min_list = temp_min_col.tolist()

print(' Coldest Min Temps (C): ',temp_min_list[0: 5], '\n Warmest Min Temps (C): ', temp_min_list[-5: -1])
```

The 5 coldest minimum temperatures in Edmonton for the year so far were during a 5 day stretch from February 7 - 11. 

I can say from personal experience that it was cold!!!

```python
import matplotlib.pyplot as plt
%matplotlib inline

sorted_data_date = data.sort_values(by='Date')

sorted_data_date.plot('Date', 'Minimum Temperature (C)', title='Daily Min Temps for 2021 in Edmonton, AB')
plt.xticks(rotation=30)
plt.show()
```
